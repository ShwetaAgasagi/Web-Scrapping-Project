Scrapped Basketball Player details from basketball website from google.

```python
from bs4 import BeautifulSoup
import requests

webpage=requests.get("https://www.basketball-reference.com/awards/slam_500_greatest.html")
webpage
```

```python
sp=BeautifulSoup(webpage.content)
sp
```

```python
table1=sp.find_all("table")
table1
```

```python
all_data=[]
for i in table1:
    for j in i.find_all("td"):
        all_data.append(j.text.replace("\n",""))
all_data
```

```python
players=all_data[0:len(all_data):15]
players
```

```python
firstyear=all_data[1:len(all_data):15]
firstyear
```

```python
lastyear=all_data[2:len(all_data):15]
lastyear
```

```python
games=all_data[3:len(all_data):15]
games
```

```python
minutesplayed=all_data[4:len(all_data):15]
minutesplayed
```

```python
pointspergame=all_data[5:len(all_data):15]
pointspergame
```

```python
totalreboundspergame=all_data[6:len(all_data):15]
totalreboundspergame
```

```python
assistpergame=all_data[7:len(all_data):15]
assistpergame
```

```python
stealspergame=all_data[8:len(all_data):15]
stealspergame
```

```python
blockspergame=all_data[9:len(all_data):15]
blockspergame
```

```python
fieldgoalpercentage=all_data[10:len(all_data):15]
fieldgoalpercentage
```

```python
point3=all_data[11:len(all_data):15]
point3
```

```python
freethrowpercentage=all_data[12:len(all_data):15]
freethrowpercentage
```

```python
winshare=all_data[13:len(all_data):15]
winshare
```

```python
winshare48min=all_data[14:len(all_data):15]
winshare48min
```

```python
import pandas as pd
df=pd.DataFrame({})
df["PLAYERS"]=players
df["FIRST_YEAR"]=firstyear
df["LAST_YEAR"]=lastyear
df["GAMES"]=games
df["MP"]=minutesplayed
df["PTS"]=pointspergame
df["TRB"]=totalreboundspergame
df["AST"]=assistpergame
df["STL"]=stealspergame
df["BLK"]=blockspergame
df["FG%"]=fieldgoalpercentage
df["3P%"]=point3
df["FT%"]=freethrowpercentage
df["WS"]=winshare
df["WS/48"]=winshare48min





df

```

```python
#Saved file to csv file

df.to_csv("basketballteam.csv")
```
