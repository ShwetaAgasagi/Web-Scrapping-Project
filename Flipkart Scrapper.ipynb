Scrapping mobile data from Flipkart website

```python
from bs4 import BeautifulSoup
import requests

page=requests.get("https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=529399fc-0b47-4e8d-a530-2ef2c3954161")
page
```

```python
sp=BeautifulSoup(page.content)
sp
```

Scrapped name,price ratings of mobile

```python
mname=sp.find_all("div",class_="_4rR01T")
mname
```

```python
price=sp.find_all("div",class_="_30jeq3 _1_WHN1")
price
```

```python
rat=sp.find_all("div",class_="_3LWZlK")
rat
```

```python
mobile1=[]
for i in mname:
    mobile1.append(i.text)
mobile1
```

```python
price1=[]
for i in price:
    price1.append(i.text)
price1
```

```python
rating1=[]
for i in rat:
    rating1.append(i.text)
rating1
```

```python
#With the help of pandas libraries creating dataframe

import pandas as pd
df=pd.DataFrame({})
df["MOBILE"]=mobile1
df["PRICE"]=price1
df["RATINGS"]=rating1[0:24]
df
```
